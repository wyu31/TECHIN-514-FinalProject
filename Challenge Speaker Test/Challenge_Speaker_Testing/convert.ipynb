```python
import wave
import numpy as np

# Open the WAV file
with wave.open('train32_v.wav', 'r') as wav_file:
    # Get the number of frames in the file
    n_frames = wav_file.getnframes()
    
    # Read the audio frames
    audio_frames = wav_file.readframes(n_frames)
    
    # Convert the frames to 16-bit samples
    samples = np.frombuffer(audio_frames, dtype=np.int16)
    
    # Normalize the samples to the range 0 to 1
    samples = samples / np.max(np.abs(samples))
    
    # Convert the samples to a 12-bit range (0 to 4095)
    samples_12bit = np.int16((samples + 1) * 2047.5)
    
    # Ensure all values are in the range 0 to 4095
    samples_12bit = np.clip(samples_12bit, 0, 4095)

# Convert to a C array
c_array = ', '.join(map(str, samples_12bit.tolist()))

# Write to a header file
with open('audio_data.h', 'w') as header_file:
    header_file.write('#ifndef AUDIO_DATA_H\n')
    header_file.write('#define AUDIO_DATA_H\n\n')
    header_file.write('const uint16_t audioData[] = {\n')
    header_file.write(c_array)
    header_file.write('\n};\n\n')
    header_file.write('#endif // AUDIO_DATA_H\n')
```
